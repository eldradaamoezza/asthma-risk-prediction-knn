##Outlier

```python
# ==============================
# 3) MEMERIKSA OUTLIER
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# LOAD DATASET
# ==============================
path = "/content/drive/MyDrive/Colab Notebooks/ASMAdataset_no_duplicates.csv"
data = pd.read_csv(path)

print("Shape dataset:", data.shape)
display(data.head())


# ==============================
# 1) Fungsi Melihat Outlier (BOXPLOT)
# ==============================
def numerical_dis_plot(features, df, segment_feature=None, showfliers=True):
    fig, ax = plt.subplots(len(features), 1, figsize=(8, 15))
    for i, feature in enumerate(features):
        if segment_feature:
            sns.boxplot(y=segment_feature, x=feature, data=df, ax=ax[i], showfliers=showfliers)
            ax[i].set_ylabel(None)
        else:
            sns.boxplot(x=feature, data=df, ax=ax[i], showfliers=showfliers)
    plt.tight_layout()
    plt.show()


# ==============================
# 2) Melihat Outlier (ANALISIS IQR UPPER-LOWER)
# ==============================
def show_outlier_description(features, df):
    print("-"*40)
    print("ANALISIS OUTLIER")
    print("-"*40)

    for feature in features:
        clean_series = pd.to_numeric(df[feature], errors='coerce').dropna()

        if clean_series.empty:
            print(f"\n--- Kolom: {feature} ---")
            print("Status: Tidak ada data numerik valid untuk dianalisis.")
            print("-"*30)
            continue

        Q1 = clean_series.quantile(0.25)
        Q2 = clean_series.median()
        Q3 = clean_series.quantile(0.75)
        IQR = Q3 - Q1

        upper_bound = Q3 + 1.5 * IQR
        lower_bound = Q1 - 1.5 * IQR

        outlier_atas = (clean_series > upper_bound).sum()
        outlier_bawah = (clean_series < lower_bound).sum()
        total_outlier = outlier_atas + outlier_bawah

        print(f"\n--- Analisis Outlier: {feature} ---")
        print(f"Q1 (Persentil 25%) : {round(float(Q1), 2)}")
        print(f"Median (Q2)        : {round(float(Q2), 2)}")
        print(f"Q3 (Persentil 75%) : {round(float(Q3), 2)}")
        print(f"IQR                : {round(float(IQR), 2)}")
        print(f"Ambang batas bawah : {round(float(lower_bound), 2)}")
        print(f"Ambang batas atas  : {round(float(upper_bound), 2)}")
        print(f"Jumlah Outlier Atas: {outlier_atas}")
        print(f"Jumlah Outlier Bawah: {outlier_bawah}")
        print(f"Total Outlier      : {total_outlier} ({(total_outlier/len(clean_series)*100):.2f}%)")
        print("-"*30)


# ==============================
# 3) FITUR NUMERIK
# ==============================
numerical_features = [
    "Age",
    "BMI",
    "Number_of_ER_Visits",
    "Peak_Expiratory_Flow",
    "FeNO_Level",
    "Medication_Adherence"
]

# ==============================
# 4) JALANKAN
# ==============================
numerical_dis_plot(numerical_features, data, segment_feature=None, showfliers=True)
show_outlier_description(numerical_features, data)


# ==============================
# 1) MENANGANI OUTLIER (METODE IQR)
# ==============================

# 1. Pilih fitur numerik yang ingin dibersihkan outlier-nya
kolom_target = [
    "Age",
    "BMI",
    "Number_of_ER_Visits",
    "Peak_Expiratory_Flow",
    "FeNO_Level",
    "Medication_Adherence"
]

# Pastikan semua kolom target bertipe numerik
for col in kolom_target:
    data[col] = pd.to_numeric(data[col], errors="coerce")

data_numeric = data[kolom_target]

# 2. Hitung Q1, Q3, dan IQR
Q1 = data_numeric.quantile(0.25)
Q3 = data_numeric.quantile(0.75)
IQR = Q3 - Q1

# 3. Tentukan batas bawah dan batas atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 4. Hapus baris yang mengandung outlier
data_clean = data[
    ~((data_numeric < lower_bound) | (data_numeric > upper_bound)).any(axis=1)
]

# 5. Output hasil pembersihan
print("="*40)
print("HASIL PENANGANAN OUTLIER")
print("="*40)

print(f"Jumlah Data Awal   : {len(data)}")
print(f"Jumlah Data Bersih : {len(data_clean)}")
print(f"Total Terhapus     : {len(data) - len(data_clean)} baris")

print("="*40)
print("Status: Outlier berhasil ditangani")
print("="*40)

# Gunakan dataset bersih untuk tahap selanjutnya
data = data_clean

# SAVE DATASET (NO OUTLIER)
data.to_csv("/content/drive/MyDrive/Colab Notebooks/ASMAdataset_no_outlier.csv", index=False)
print("Dataset tanpa outlier berhasil disimpan.")

```

##Visualisasi Fitur

```python
# ==============================
# 5) VISUALISASI FITUR BERDASARKAN TARGET (STACKED)
# ==============================

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
path = "/content/drive/MyDrive/Colab Notebooks/asthma_dataset.csv"
data = pd.read_csv(path, sep=';')


# Daftar fitur kategorikal yang ingin divisualisasikan
categorical_features = [
    "Gender","Smoking_Status","Family_History","Allergies",
    "Physical_Activity_Level","Occupation_Type","Comorbidities","Medication_Adherence"
]

# Kolom target
target_col = "Has_Asthma"


# Fungsi visualisasi stacked bar
def categorical_vs_target_stacked(features, target, df, max_unique=8):
    for col in features:

        s = df[col]  # Ambil data fitur

        # Jika kategori terlalu banyak  ubah menjadi 3 kelompok (Low, Medium, High)
        if s.nunique(dropna=True) > max_unique:
            s = pd.qcut(pd.to_numeric(s, errors="coerce"),
                        3,
                        labels=["Low","Medium","High"],
                        duplicates="drop")

        # Membuat stacked bar chart
        ax = pd.crosstab(df[target], s).plot(
            kind="bar",
            stacked=True,
            figsize=(7,4)
        )

        # Menampilkan angka jumlah pada setiap bar
        for c in ax.containers:
            ax.bar_label(c, label_type="center", fontsize=9)

        # Pengaturan judul dan label grafik
        plt.title(f"{col} berdasarkan {target}")
        plt.xlabel(target)
        plt.ylabel("Jumlah")
        plt.legend(title=col, bbox_to_anchor=(1.02,1), loc="upper left")

        plt.tight_layout()
        plt.show()


# Menjalankan fungsi visualisasi
categorical_vs_target_stacked(categorical_features, target_col, data)
```

##Hyperparameter Tuning

```python
# ==============================
# TUNING KNN (CV 5-FOLD) -
# ==============================

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report

X_train = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/X_train.csv")
X_test  = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/X_test.csv")
y_train = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/y_train.csv").values.ravel()
y_test  = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/y_test.csv").values.ravel()

k_values = [1,3,5,7,9,11,13,15]

# 1) Cross Validation 5-Fold
cv_scores = [cross_val_score(KNeighborsClassifier(n_neighbors=k),
                             X_train, y_train, cv=5, scoring="accuracy").mean()
             for k in k_values]

# 2) Buat tabel hasil CV
hasil_cv = pd.DataFrame({
    "Nilai_K": k_values,
    "CV_Accuracy": [round(float(s),4) for s in cv_scores]
}).sort_values("CV_Accuracy", ascending=False).reset_index(drop=True)

best_k  = int(hasil_cv.loc[0,"Nilai_K"])
best_cv = float(hasil_cv.loc[0,"CV_Accuracy"])

print("="*45)
print("HASIL TUNING KNN (CV 5-FOLD)")
print("="*45)
display(hasil_cv)
print(f"Best K           : {best_k}")
print(f"Best CV Accuracy : {best_cv:.4f}")
print("="*45)

# 3) Latih model terbaik
best_knn_model = KNeighborsClassifier(n_neighbors=best_k).fit(X_train, y_train)

print("\nTesting Accuracy :", round(accuracy_score(y_test, y_pred),4))
```
